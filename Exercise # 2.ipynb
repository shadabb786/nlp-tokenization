```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
with open('/content/drive/MyDrive/Text.txt', 'r') as file:
    text = file.read()
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=set(STOPWORDS),
    min_font_size=10
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np

# Load an image for the mask
mask = np.array(Image.open('/content/drive/MyDrive/Mask.png'))

# Generate the word cloud with a mask
wordcloud = WordCloud(
    mask=mask,
    background_color='white',
    stopwords=set(STOPWORDS)
).generate(text)
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='blue',
    stopwords=set(STOPWORDS),
    min_font_size=10
).generate(text)




plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
